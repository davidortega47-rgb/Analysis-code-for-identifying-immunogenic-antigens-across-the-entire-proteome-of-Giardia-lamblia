## File analysis
This notebooks does the exploratory data analysis of the Combined file (i.e. GL50581_Combined.csv) generated in the file_merge.ipynb notebook

```python
import pandas as pd

path = '../LIBCE_RESULTS/RESULTS-ProteomaGS-MOUSE-19-12-2025/GL50581_Combined.csv'
df = pd.read_csv(path) 


print("shape:", df.shape)
display(df.head())
```

```python
cols = ['Protein Length', 'Molecular Weight']
null_counts = df[cols].isnull().sum()
print("Null counts:")
print(null_counts)
```

```python
filtered_df_5 = df[(df['rank'] <= 5) & (df['Protein Length'] <= 3000) & (df['ic50'] <= 50)]

print("Number of peptides", filtered_df_5.shape[0])
print("Unique proteins", filtered_df_5['protein'].nunique())
```

```python
filtered_df_1 = df[(df['rank'] <= 1) & (df['Protein Length'] <= 3000) & (df['ic50'] <= 50)]

print("Number of peptides", filtered_df_1.shape[0])

print("Unique proteins", filtered_df_1['protein'].nunique())
```

```python
filtered_df_01 = df[(df['rank'] <= 0.1) & (df['Protein Length'] <= 3000) & (df['ic50'] <= 50)]

print("Number of peptides", filtered_df_01.shape[0])

print("Unique proteins", filtered_df_01['protein'].nunique())
```

```python
filtered_df_001 = df[(df['rank'] <= 0.01) & (df['Protein Length'] <= 3000) & (df['ic50'] <= 50)]

print("Number of peptides", filtered_df_001.shape[0])

print("Unique proteins", filtered_df_001['protein'].nunique())
```

```python
cyst_df = filtered_df_001[filtered_df_001['Cyst'] == 1]
unique_gene_count = cyst_df['Gene ID'].nunique()
total_rows = cyst_df.shape[0]

print(f"Unique Cyst proteins with peptides with rank <= 0.01: {unique_gene_count}")
print(f"Total number of peptides : {total_rows}")
```

```python
trophozoite_df = filtered_df_001[filtered_df_001['Trophozoite'] == 1]
unique_gene_count = trophozoite_df['Gene ID'].nunique()
total_rows = trophozoite_df.shape[0]

print(f"Unique Trophozoite proteins with peptides with rank <= 0.01: {unique_gene_count}")
print(f"Total number of peptides : {total_rows}")
```

```python
secretome_df = filtered_df_001[filtered_df_001['Secretome'] == 1]
unique_gene_count = secretome_df['Gene ID'].nunique()
total_rows = secretome_df.shape[0]

print(f"Unique Secretome proteins with peptides with rank <= 0.01: {unique_gene_count}")
print(f"Total number of peptides : {total_rows}")
```

```python

cyst_count = int((filtered_df_001['Cyst'] == 1).sum())
tropho_count = int((filtered_df_001['Trophozoite'] == 1).sum())
secretome_count = int((filtered_df_001['Secretome'] == 1).sum())
tm_domains_count = int((filtered_df_001['# TM Domains'] >= 1).sum())

print(f"Cyst == 1: {cyst_count}")
print(f"Trophozoite == 1: {tropho_count}")
print(f"Secretome == 1: {secretome_count}")
print(f"# TM Domains >= 1: {tm_domains_count}")
```

## Allele analysis

```python
grouped = (
    filtered_df_001
    .groupby('allele')
    .agg(peptides=('allele', 'size'), proteins=('Gene ID', 'nunique'))
    .reset_index()
    .sort_values('peptides', ascending=False)
)
display(grouped)
```

## Promiscuous Peptide

```python
filtered_df_001
```

```python
promiscuous_peptides = (
    filtered_df_001
    .groupby('peptide')['allele']
    .apply(lambda x: x.nunique())
    .reset_index()
    .rename(columns={'allele': 'unique_alleles'})
)
promiscuous_peptides = promiscuous_peptides[promiscuous_peptides['unique_alleles'] > 1]
promiscuous_peptides['count'] = promiscuous_peptides['peptide'].map(
    filtered_df_001['peptide'].value_counts()
)
promiscuous_peptides = promiscuous_peptides.sort_values('count', ascending=False)
display(promiscuous_peptides)
```
